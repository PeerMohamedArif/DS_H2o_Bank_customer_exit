#### Printing the categorical variables

```python
def print_unique_col_values(dataset):
    for column in dataset:
        if dataset[column].dtypes=='object':
            print(f'{column} : {dataset[column].unique()}')

print_unique_col_values(dataset)
```

# Label Encoding

```python
dataset['Gender'].replace({'Male':1,'Female':0},inplace=True) # here they have used dictionary you can also use list where 0 index will have all old values and 
#1 index will have new values in the replace function
```

```python
dataset.head()
```

```python
dataset['Geography'].replace({'France':'Fr','Spain':'Sp','Germany':'Ge'},inplace=True)
```

```python
dataset.head()
```

```python
# ONE Hot Encoding Method
```

```python
dataset1 = pd.get_dummies(data=dataset, columns=['Geography'])
dataset1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaled_variable = ['Tenure','CreditScore','Age','Balance','NumOfProducts','EstimatedSalary']
scaler=MinMaxScaler()
dataset1[scaled_variable]=scaler.fit_transform(dataset[scaled_variable])
```

```python
dataset1.head()
```

```python
X = dataset1.drop('Exited',axis='columns')  ##independent features
y = dataset1['Exited']  ##dependent feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5) #Consistency Across Runs:

#When you set random_state to a specific integer (e.g., 5 in your example), it initializes the random number generator in a fixed way. 
#This means that every time you run the code, you will get the same split of X and y into X_train, X_test, y_train, and y_test.
#If random_state is not set, each run could produce a different split, 
#which may lead to variations in model performance due to different training/testing data.
```

```python
len(X_train.columns) # ytrain is one column
```

## Model Building and Prediction

## The Sequential model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    keras.layers.Dense(12, input_shape=(14,), activation='relu'),
    keras.layers.Dense(6, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])


model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
pred = model.predict(X_test)
pred
```

```python
y_pred=[]
for element in pred:
    if element>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred
```

# Accuracy Check

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

mplt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True, fmt='d')
mplt.xlabel('Predicted')
mplt.ylabel('Truth')
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy score is: ", accuracy_score(y_test,y_pred)*100,"%")
```
